1. Part 1.1 - Denoising: Gaussian Filter


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# image filtering
image_path = "./gaussian_noise_peppers.png"
img = cv2.imread(image_path)

# filter image using 7x7 Gaussian, the std is calculated automatically
res1 = cv2.GaussianBlur(img,(7,7), np.sqrt(2))

# convert from BGR to RGB for displaying
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res1 = cv2.cvtColor(res1, cv2.COLOR_BGR2RGB)

# plot
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(img)
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(res1)
plt.title("7x7 Gaussian"), plt.xticks([]), plt.yticks([])
plt.show()
```

2. Part 1.1 - Denoising: Box Filter


```python
# image filtering
image_path = "./gaussian_noise_peppers.png"
img = cv2.imread(image_path)

# perform box blur
res1 = cv2.blur(img, (7,7))

# convert from BGR to RGB for displaying
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res1 = cv2.cvtColor(res1, cv2.COLOR_BGR2RGB)

# plot
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(img)
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(res1)
plt.title("Box filter"), plt.xticks([]), plt.yticks([])
plt.show()
```

3. Part 1 - Denoising: Compare the PSNR


```python
import skimage
from skimage import measure

# Load the images
image_path1 = "./peppers.png"
image_path2 = "./gaussian_noise_peppers.png"
img_true = cv2.imread(image_path1)
img_gausNoise = cv2.imread(image_path2)

# Calculate the PSNR for gaussian and box blur
img_gaussian = cv2.GaussianBlur(img_gausNoise, (7,7), np.sqrt(2))
img_box = cv2.blur(img_gausNoise, (7,7))
gaussianPSNR = skimage.measure.compare_psnr(img_true, img_gaussian)
boxPSNR = skimage.measure.compare_psnr(img_true, img_box)

print("gaussianPSNR:",gaussianPSNR)
print("boxPSNR:",boxPSNR)
```

Compare the PSNR: since PSNR is the peak signal to noise ratio, the method with a higher PSNR should output images with higher quality. Therefore, the gaussian blur gives better result in this case.

5. Part 1 - Denoising: Median Filter for Salt and Pepper

```python
# image filtering
image_path = "./saltandpepper_noise_peppers.png"
img = cv2.imread(image_path)

# filter image using 5x5 median filter
res = cv2.medianBlur(img, 7)

# convert from BGR to RGB for displaying
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)

# plot
plt.figure(figsize=(20, 20))
plt.subplot(121), plt.imshow(img)
plt.title("Image"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(res)
plt.title("Median"), plt.xticks([]), plt.yticks([])
plt.show()
```

6. Part 1 - Denoising: Compare PSNR for Salt and Pepper

```python
import skimage
from skimage import measure

image_path1 = "./peppers.png"
img_true = cv2.imread(image_path1)
image_path2 = "./saltandpepper_noise_peppers.png"
img_saltAndPepper = cv2.imread(image_path2)

img_gaussian = cv2.GaussianBlur(img_saltAndPepper, (7,7), np.sqrt(2))
img_median = cv2.medianBlur(img_saltAndPepper, 7)
gaussianPSNR = skimage.measure.compare_psnr(img_true, img_gaussian)
medianPSNR = skimage.measure.compare_psnr(img_true, img_median)

print("gaussianPSNR:",gaussianPSNR)
print("medianPSNR:",medianPSNR)
```

Compare the PSNR: since PSNR is the peak signal to noise ratio, the method with a higher PSNR should output images with higher quality. Therefore, the median blur method gives better result in this case.
