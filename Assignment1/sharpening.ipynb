1. Part 2 - Sharpening: Sharpen the image "rice"

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = "./rice.png"
img_rice = cv2.imread(img_path)

#generate the blurred images for offset calculation
img_box = cv2.blur(img_rice,(5,5))
img_gaussian = cv2.GaussianBlur(img_rice,(5,5), np.sqrt(2))

#details = original image - blurred image
#sharpened image = original image + details

#Calculate the sharpened image using box filtered image
detail_box = img_rice - img_box
sharpened_box = img_rice + detail_box

#Calculate the sharpened image using gaussian filtered image
detail_gaussian = img_rice - img_gaussian
sharpened_gaussian = img_rice + detail_gaussian

#Convert the color of the sharpened images for display
sharpened_box = cv2.cvtColor(sharpened_box, cv2.COLOR_BGR2RGB)
sharpened_gaussian = cv2.cvtColor(sharpened_gaussian, cv2.COLOR_BGR2RGB)

# plot the figures
plt.figure(figsize=(20, 20))
plt.subplot(131), plt.imshow(img_rice)
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(sharpened_box)
plt.title("Sharpened-Box"), plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(sharpened_gaussian)
plt.title("Sharpened-Gaussian"), plt.xticks([]), plt.yticks([])
plt.show()
```

2. Part 2 - Sharpening: Which of the two methods is expected to give better results and why? Can you observe expected result?

The box sharpening method should give out better result. The box method computes the average of the surrounding pixels of the current pixel, while the gaussian blurring method weighs the contribution of neighbouring pixels by nearness. This would cause the gaussian blurred image to be more similar to the original image than the box blurred image.

The details is calculated using blurred image subtracted from the original image and the sharpened image is obtained by adding details to original image. As a result, the box blurring method would give out a higher details image than gaussian blurring method, thus having a higher offset to add to the original image to compute the sharpened image. Therefore, the box blurring method outputs a better sharpened image.

And yes, we can observe the expected result from the images since the box-sharpened image has more crisp edges than the gaussian-sharpened image
